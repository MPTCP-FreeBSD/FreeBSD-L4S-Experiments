```python
import dpkt
import os
import pandas as pd

global dataframe

def calculate_total_throughput(pcap_file):
    total_bytes = 0
    with open(pcap_file, 'rb') as f:
        pcap = dpkt.pcap.Reader(f)
        for timestamp, buf in pcap:
            # Assuming Ethernet frames, you may need to adjust if working with other protocols
            eth = dpkt.ethernet.Ethernet(buf)
            total_bytes += len(eth)
    return total_bytes / (1024 * 1024)

def main():
    # Directory containing pcap files
    pcap_folder = './datatest2'

    # CSV file to store results
    csv_file = 'pcap_throughput.csv'

    data = []

    for file_name in os.listdir(pcap_folder):
        if file_name.endswith('.pcap'):
            file_path = os.path.join(pcap_folder, file_name)
            total_throughput = calculate_total_throughput(file_path)
            if total_throughput != 0:
                data.append({'File Name': file_name, 'Total Throughput (bytes)': total_throughput})

    df = pd.DataFrame(data)
    global dataframe
    dataframe = df.copy()
    print(df)
    df.to_csv(csv_file, index=False)

    print("Total throughput calculated and saved to", csv_file)

if __name__ == "__main__":
    main()
```
